```python
# Extracts target images from original fMoW test data set
import os
import shutil
import glob
```

```python
#*********************************************************************
# ***********************80/20 Train/Val Split***********************

splitlist = ['/Users/johndurkee/Training_data/cropped_images/train/oil_gas_fac',
             '/Users/johndurkee/Training_data/cropped_images/train/storage_tank',
             '/Users/johndurkee/Training_data/cropped_images/train/tunnel',
             '/Users/johndurkee/Training_data/cropped_images/train/helipad',
             '/Users/johndurkee/Training_data/cropped_images/train/tg_non_tgt',
             '/Users/johndurkee/Training_data/new_cropped_images/train/helipad'
          ]

val_dir = ['/Users/johndurkee/Training_data/cropped_images/val/oil_gas_fac',
           '/Users/johndurkee/Training_data/cropped_images/val/storage_tank',
           '/Users/johndurkee/Training_data/cropped_images/val/tunnel',
           '/Users/johndurkee/Training_data/cropped_images/val/helipad',
           '/Users/johndurkee/Training_data/cropped_images/val/v_non_tgt',
           '/Users/johndurkee/Training_data/new_cropped_images/val/helipad'
          ]

path = splitlist[5] # Need to manually index splitlist here
train = glob.glob((path + '/*.jpg'))
print(f'Total in training folder: {len(train)}')
val_split = int(0.2*len(train))
print(f'val_split = {val_split}')

valfile_list = train[-val_split:]
print(f'{len(valfile_list)} images sectioned off for val set')
```

```python
i = 0
for f in valfile_list:
    i += 1
    path = splitlist[4].split('/')[-2] # Need to manually index splitlist here
    file_ = f.split('/')[-1]
    print(file_)
    path = os.path.join(val_dir[4], file_)# Need to manually index val_dir here
    print(path)
    print('=================================================')
    shutil.move(f, path)
print(f'{i} files moved to validation folder')
```
