```python
import random
import os
import shutil
import glob
```

```python
# Randomly select folder names to choose non-target classes
# from original test data set

a = 7270 # no. of largest tgt class files in training/validation sets
b = 866 # no. of largest tgt class files in test set 
x = a + b
n = x

#Generate n random numbers between 0 and x
randomlist = random.sample(range(0, x), n)
print(len(randomlist))
print(len(set(randomlist)))
trainlist = randomlist[:b]
testlist = randomlist[b:]
print(len(trainlist))
print(trainlist[:10])
print(len(testlist))
print(testlist[:10])
```

```python
# Generate list of folders remaining in orginal test set
from_dir = '/Users/johndurkee/Training_data/cropped_images/test/test_non_tgt_only'
from_files = glob.glob(from_dir + '/*_rgb.jpg')
folders = []
for f in from_files:
    f = f.split('/')[-1]
    f = f.split('_')[-3]
    folders.append(f)
folders = set(folders)
folders = list(folders)
print(f'{len(folders)} folders in: {from_dir}')
print(folders[:10])
```

```python
# Pass random lists for non-tgt class **training set** through entire list of test data
# This will create a list of random folders that can be used for **training set**
test_from_test = []
for num in trainlist:
    temp_place = folders[num]
    test_from_test.append(temp_place)
print(len(test_from_test))
print(test_from_test[:10])
```

```python
# Pass random lists for non-tgt class **test set** through entire list of test data
# This will create a list of random folders that can be used for **test set**
train_from_test = []
for num in testlist:
    temp_place = folders[num]
    train_from_test.append(temp_place)
print(len(train_from_test))
print(train_from_test[:10])
```

```python
# STEP 1
# Here is the one for the Last SET O FILES TO COPY (REALY NEED TO MOVE THIS)
from_dir = '/Users/johndurkee/Training_data/cropped_images/test/test_non_tgt_only'
from_files = glob.glob(from_dir + '/*_rgb.jpg')
print(f'{len(from_files)} images in: {from_dir}')
to_dir = '/Users/johndurkee/Training_data/cropped_images/test/non_tgt'
```

```python
# STEP 2
# Moves all the target files from original target list to new directory
# Takes 2 Args
# Arg 1: file map of tgt classes: all helipad foldernames
# Arg 2: list of all original test files (or train non target files)
def copy_tgt_files(class_list, from_files):
    i = 0
    out_list = []
    for name in class_list: #Tunnel_list
        for test_name in from_files: #Entire List of files in Test Folder (From original test set)
          if name == test_name.split('.')[0].split('_')[-3].split('/')[-1]:
            i += 1
            new_path = os.path.join(to_dir, test_name.split('/')[-1])
            out_list.append(new_path)
            shutil.move(test_name, new_path)
    print(f'Transfer files complete. {i} files moved here: {to_dir}')
    print(f'Should have {len(from_files) - i} remaining in: {from_dir}')
    return out_list
```

```python
# Extract random training non-tgt class images from test set
# Extract random test targets from test set

#random_train_non_tgt_class_images = copy_tgt_files(train_from_test, from_files)
random_test_non_tgt_class_images = copy_tgt_files(test_from_test, from_files)
```
